```python
def Matrix_chain_order(p):
    """
    Minimization the number of operation for the multiplication of matrices
    p: list, size of each matrix
"""
    n=len(p)-1 # the number of matrixes
    # the size of Ai matrix is p_i*p_{i+1}
    m=np.zeros([n,n])
    s=np.zeros([n-1,n])
    for l in range(2,n+1):
        for i in range(0,n-l+1):
            j=l+i-1
            m[i,j]=np.inf
            for k in range(i,j):
                q=m[i,k]+m[k+1,j]+p[i]*p[k+1]*p[j+1]
                if q<m[i,j]:
                    m[i,j]=q
                    s[i,j]=k
    return m,s
```

```python
def Extended_bottom_up_cut_rod(p,n):
    """
    Maximize the price of cutting rod, given the length of each rod p_i for the length i and total length n
    p: list, length n
    n: the total length
"""
    r = [0 for _ in range(n+1)] # the price for each length
    s = [0 for _ in range(n+1)]
    for j in range(1,n+1): # the range of length is 1 to n
        q=-float('inf')
        for i in range(1,j+1): # the range of the first part is 1 to j
#             print(i,j-i)
            if q< p[i-1]+r[j-i]:#找到更大的，该长度对应的节点
                q=p[i-1]+r[j-i]
                s[j]=i
        r[j]=q
    return r,s

r,s=Extended_bottom_up_cut_rod([1,5,8,9,10,17,17,20,24,30],10)
print('Gain for different length',r)
print('cut point',s)
print('scheme for n=7')
n=7
scheme=[]
n=int(n)
while n!=0:
    scheme.append(s[int(n)])
    n=n-s[n]
print(scheme)
```

```python
# 413. Arithmetic Slices，给定array里等差子数列的数量
def numberOfArithmeticSlices(A):
    """
    :type A: List[int]
    :rtype: int
    """
    B = [A[i+1]-A[i] for i in range(len(A)-1)]
    res = 0
    for i in range(1,len(B)):
        start = i-1
        while start>=0 and B[start] == B[i]:
            start-=1
        res+=i-start-1
    return res
numberOfArithmeticSlices(A = [1, 2, 3, 4])
```

```python
# 338. Counting Bits, 0-num之间每个数的bit中1的个数
# 利用个数每翻一倍，在最前面增加1
def countBits(num):
    """
    :type num: int
    :rtype: List[int]
    """
    res = [0,1]
    if num<2:
        return res[:num+1]
    while len(res)<=num+1:
        res = res+[i+1 for i in res]
    return res[:num+1]
countBits(5)
```

```python
# 22. Generate Parentheses
# 剩下n个(和n个)，如果left<right，那么可以随便插入，如果left=right，那么必须插入(,如果剩下的数都为0，那么完成
# 学会做这种有多个返回的情况，dfs中一个参量为ans，一个参量为流动解，在达到某种情况下，流动解插入到ans中
# dp方法，Fn=sum(Fi+F[n-i])，修改一下
def generateParenthesis(n):
    """
    :type n: int
    :rtype: List[str]
    """
    def dfs(left,right,path,res):
        if left==0 and right==0:
            res.append(path)
        elif left>=0:
            if left<=right:
                dfs(left-1,right,path+'(',res)
            if left<right:
                dfs(left,right-1,path+')',res)

    res = []
    dfs(n,n,"",res)
    return res
    
def generateParenthesis(n):
    """
    :type n: int
    :rtype: List[str]
    """
    dp = [[] for i in range(n + 1)]
    dp[0].append('')
    for i in range(n + 1):
        for j in range(i):
            dp[i] += ['(' + x + ')' + y for x in dp[j] for y in dp[i - j - 1]]
    return dp[n]
generateParenthesis(2)
```

```python
# 132. Palindrome Partitioning II
# 类似于coin change，当前回文和对应的fi，但这个是O3复杂度
# O2复杂度，避免不断的reverse比较而是用元素比较，类比最长回文子串
# low-high为当前回文子串的index（相当于对应的回文区域），cut[low]为前面的fi，cut[high+1]为high所在的index，这样不断更新
def minCut(s):
    """
    :type s: str
    :rtype: int
    """
    dp = [i-1 for i in range(len(s)+1)]

    def isPalindrome(s):
        return s==s[::-1]

    for i in range(len(s)):
        for j in range(i+1):
            if isPalindrome(s[j:i+1]):
                dp[i+1] = min(dp[i+1],dp[j]+1)
    return dp[-1]

def minCut(s):
    """
    :type s: str
    :rtype: int
    """
    cut = [i-1 for i in range(len(s)+1)]
    for idx in range(1, len(s)):
        for low, high in (idx, idx), (idx - 1, idx):
            while low >= 0 and high < len(s) and s[low] == s[high]:
                cut[high + 1] = min(cut[high + 1], cut[low] + 1)
                low -= 1
                high += 1
    return cut[-1]
    
minCut('aab')
```

## 背包knapsack问题
$F(i,j)$为前i个商品在重量小于等于j的情况下最大价值<br/>
$ F(i,j) = max\{F(i-1,j), v_i + F(i-1,j-w_i)\},j-w_i>=0; F(i-1,j), j-w_i<0$

```python
def knapsack(v,w,weight):
    n = len(v)
    dp = [[0 for _ in range(weight+1)] for _ in range(n+1)]#前i个物品，前j个重量下的最大值
    for i in range(n):
        for j in range(1,weight+1):
            if w[i] <= j:#如果当前物品小于总重量，那么试试带走
                dp[i+1][j] = max(dp[i][j-w[i]]+v[i],dp[i][j])
            else:#否则不带走
                dp[i+1][j] = dp[i][j]
    return dp[-1][-1]

# if __name__ == '__main__':
v = [60,100,120]
w = [10,20,30]
weight = 50
knapsack(v,w,weight)
```

```python
# 474. Ones and Zeroes
# 无value的01背包问题，给定0和1的数量，和多个备选项，问最多能找出多少个
def findMaxForm(strs, m, n):
    """
    :type strs: List[str]
    :type m: int
    :type n: int
    :rtype: int
    """
    dp = [[0 for _ in range(n+1)] for _ in range(m+1)]
    for t in range(len(strs)):
        newdp = [[0 for _ in range(n+1)] for _ in range(m+1)]
        count0 = strs[t].count('0')
        count1 = len(strs[t]) - count0
        for i in range(m+1):#the number of 0s
            for j in range(n+1):
                newdp[i][j] = dp[i][j]
                if count0<=i and count1<=j:
                    newdp[i][j] = max(newdp[i][j],dp[i-count0][j-count1]+1)
        dp = newdp
    return dp[-1][-1]
findMaxForm(["01","0", "1"], m = 1, n = 1)
```

无限背包问题<br/>
$dp[i][v] = max\{dp[i-1][v - k * c[i]] + k * w[i] | 0 <= k * c[i]<= v\}$

```python
def knapsack2(v,w,weight):
    """完全背包问题"""
    n = len(v)
    dp = [0 for _ in range(weight+1)]
    for i in range(n):
        for j in range(1,weight+1):
            if w[i] <= j:
                dp[j] = max(dp[j],dp[j-w[i]]+v[i])
    return dp[-1]

# if __name__ == '__main__':
v = [60,100,120]
w = [10,20,30]
weight = 50
result = knapsack2(v,w,weight)
print(result)
```

```python
#805. Split Array With Same Average，将数组分成两部分，平均值相同
import bisect
def splitArraySameAverage(A):
    """
    :type A: List[int]
    :rtype: bool
    """
    if len(A)==1:
        return False
    sums,lens = sum(A),len(A)
    # A[k]/k=sums/lens，A中选取k个数，使得和为0
    res = sorted([lens*a-sums for a in A])
    if 0 in res:
        return True
    neg = res[:bisect.bisect_left(res,0)]
    pos = res[bisect.bisect_left(res,0):]
    # 从neg中挑选一部分数字，从pos中挑选一部分数字，构成0
    if len(neg)>len(pos):
        neg,pos = pos,neg
    possible_sums = {0}
    for n in neg:
        possible_sums.update({(v + n) for v in possible_sums})
    possible_sums.remove(0)
    possible_sums.remove(sum(neg))

    level = {0}
    for i in pos:
        newlevel = set()
        for node in level:
            t = node+i
            if -t in possible_sums:
                return True
            newlevel.add(t)
        level.update(newlevel)
    return False
splitArraySameAverage([10,29,13,53,33,48,76,70,5,5])
```

```python
# 1563. Stone Game V，memorized dp分成左右两堆
def stoneGameV(stoneValue):
    """
    :type stoneValue: List[int]
    :rtype: int
    """
    presum = [0]
    for i in stoneValue:
        presum.append(presum[-1]+i)
    dp = dict()

    def dfs(left,right):
        if (left,right) in dp:
            return dp[(left,right)]
        curtotal = presum[right+1]-presum[left]
        curres = 0
        for i in range(left,right):#i及其左边的
            cur = presum[i+1]-presum[left]
            if cur<curtotal-cur:
                curres = max(curres,dfs(left,i)+cur)
            elif cur>curtotal-cur:
                curres = max(curres,dfs(i+1,right)+curtotal-cur)
            else:
                curres = max(curres,dfs(left,i)+cur,dfs(i+1,right)+curtotal-cur)
        dp[(left,right)] = curres
        return curres
    return dfs(0,len(stoneValue)-1)

stoneGameV(stoneValue = [6,2,3,4,5,5])
```

```python
# 1043. Partition Array for Maximum Sum,每段长度最多K，每段的值为最大值乘以个数，求最大和
# Time O(NK), Space O(N)
def maxSumAfterPartitioning(A, K):
    """
    :type A: List[int]
    :type K: int
    :rtype: int
    """
    dp = dict()#从index开始有划分的最大值

    def dfs(index):
        if index==len(A):
            return 0
        if index in dp:
            return dp[index]

        curmax = A[index]
        maximum = curmax + dfs(index+1)
        for i in range(index+1,min(index+K,len(A))):
            curmax = max(curmax,A[i])
            maximum = max(maximum,curmax*(i-index+1)+dfs(i+1))
        dp[index] = maximum
        return maximum

    return dfs(0)

def maxSumAfterPartitioning(A, K):
    N = len(A)
    dp = [0] * (N + 1)
    for i in range(N):
        curMax = 0
        for k in range(1, min(K, i + 1) + 1):
            curMax = max(curMax, A[i - k + 1])
            dp[i] = max(dp[i], dp[i - k] + curMax * k)
    return dp[N - 1]

maxSumAfterPartitioning(A = [1,15,7,9,2,5,10], K = 3)
```

```python
# 1463. Cherry Pickup II
def cherryPickup(grid):
    """
    :type grid: List[List[int]]
    :rtype: int
    """
    m,n = len(grid),len(grid[0])
    dp = dict()

    def dfs(index,c1,c2):
        if (index,c1,c2) in dp:
            return dp[(index,c1,c2)]
        if index==m:
            return 0
        res = 0
        for x1 in [c1-1,c1,c1+1]:
            for x2 in [c2-1,c2,c2+1]:
                if x1!=x2 and 0<=x1<n and 0<=x2<n:
                    res = max(res,dfs(index+1,x1,x2)+grid[index][c1]+grid[index][c2])
        dp[(index,c1,c2)] = res
        return res

    dfs(0,0,n-1)
    return max(dp.values())
cherryPickup(grid = [[3,1,1],[2,5,1],[1,5,5],[2,1,1]])
```

```python
# 329. Longest Increasing Path in a Matrix，matrix里最长的路径长度
def longestIncreasingPath(matrix):
    """
    :type matrix: List[List[int]]
    :rtype: int
    """
    if not matrix or not matrix[0]:
        return 0
    m,n = len(matrix),len(matrix[0])
    dp = dict()
    maximum = [0]


    def dfs(i,j):
        if (i,j) in dp:
            return dp[(i,j)]
        res = 1
        for dx,dy in [(-1,0),(1,0),(0,1),(0,-1)]:
            x,y = i+dx,j+dy
            if 0<=x<m and 0<=y<n and matrix[x][y]>matrix[i][j]:
                res = max(res,dfs(x,y)+1)
        dp[(i,j)] = res
        maximum[0] = max(maximum[0],res)
        return res

    for i in range(m):
        for j in range(n):
            dfs(i,j)
    return maximum[0]

longestIncreasingPath([
  [9,9,4],
  [6,6,8],
  [2,1,1]
])
```

```python
# 1125. Smallest Sufficient Team

def smallestSufficientTeam(req_skills, people):
    """
    :type req_skills: List[str]
    :type people: List[List[str]]
    :rtype: List[int]
    """
    dp = dict()
    dic = dict()
    for i,j in enumerate(req_skills):
        dic[j] = i
    n = len(req_skills)

    def dfs(index,skills):#从index开始，已经有了skills情况下输出的答案
        if (index,skills) in dp:
            return dp[(index,skills)]
        if skills+1 == 1<<n:#如果全部满足了，那么不需要更多的
            return 0,[]
        if index==len(people):#如果人走完了
            return float('inf'),[]

        newskills = skills
        for i in people[index]:#包含当前的情况
            if i in dic:
                newskills = newskills | 1<<dic[i]
        if skills==newskills:
            dp[(index,skills)] = dfs(index+1,skills)
        else:
            r1 = dfs(index+1,skills)#不要当前的人
            r2 = dfs(index+1,newskills)#要当前的人
            if r1[0]<1+r2[0]:
                dp[(index,skills)] = r1
            else:
                dp[(index,skills)] = [r2[0]+1,[index]+r2[1]]
        return dp[(index,skills)]

    return dfs(0,0)[1]

def smallestSufficientTeam(req_skills, people):# Time O(people * 2^skill)
    n, m = len(req_skills), len(people)
    key = {v: i for i, v in enumerate(req_skills)}
    dp = {0: []}##所拥有的skills需要的最少人数
    for i, p in enumerate(people):
        his_skill = 0#每一个人的能力
        for skill in p:
            if skill in key:
                his_skill |= 1 << key[skill]
        for skill_set, need in list(dp.items()):
            with_him = skill_set | his_skill
            if with_him == skill_set: continue#如果这个人有没有都一样，那么就不用管
            if with_him not in dp or len(dp[with_him]) > len(need) + 1:#如果这个人增加了新能力，或者这个人减少了需要
                dp[with_him] = need + [i]
    return dp[(1 << n) - 1]
smallestSufficientTeam(req_skills = ["algorithms","math","java","reactjs","csharp","aws"], people = [["algorithms","math","java"],["algorithms","math","reactjs"],["java","csharp","aws"],["reactjs","csharp"],["csharp","math"],["aws","java"]])
```

```python
# 847. Shortest Path Visiting All Nodes,给定一个图，所有的edge都是1，问从任何一点出发，走完全程的最短步数
def shortestPathLength(graph):
    """
    :type graph: List[List[int]]
    :rtype: int
    """
    queue = [(i,1<<i) for i in range(len(graph))]#当前节点，与当前走过的节点，从任何一点出发都可以
    seen = {(i,1<<i) for i in range(len(graph))}
    step = 0
    while queue:
        newqueue = []
        for node,visited in queue:
            if visited+1 == 1<<len(graph):
                return step
            for newnode in graph[node]:#往前走一步
                if (newnode,visited|(1<<newnode)) not in seen:
                    newqueue.append((newnode,visited|(1<<newnode)))
                    seen.add((newnode,visited|(1<<newnode)))
        queue = newqueue
        step+=1
```

```python
# LCP 13. 寻宝
# 给定起点S和终点T，中途需要完成OM循环，问最少步数
# 只有4种路径，SO,OM,MO,MT，将起点视为一关，需要走完一个个M，首先预先计算出所有M之间的相互距离，问题就变成了从一个地方出发，途径所有地方，
# 最短距离到达终点的旅行商问题，然后用bitmask
def minimalSteps(maze):
    """
    :type maze: List[str]
    :rtype: int
    """
    m,n = len(maze),len(maze[0])
    stones = []
    machines = []
    for i in range(m):
        for j in range(n):
            if maze[i][j]=="S":
                start = (i,j)
            if maze[i][j] == 'T':
                target = (i,j)
            if maze[i][j] == "M":
                machines.append((i,j))
            if maze[i][j] == 'O':
                stones.append((i,j))
    machines = [start]+machines
    
    def calc_dist(i,j):
        arr = [[float('inf') for _ in range(n)] for _ in range(m)]
        arr[i][j] = 0
        step = 1
        queue = [(i,j)]
        while queue:
            newqueue = []
            for i,j in queue:
                for dx,dy in [(-1,0),(1,0),(0,-1),(0,1)]:
                    x,y = i+dx,j+dy
                    if 0<=x<m and 0<=y<n and maze[x][y]!='#' and arr[x][y]==float('inf'):
                        arr[x][y] = step#要在这里更新，否则的话会添加多个(x,y)
                        newqueue.append((x,y))
            step+=1
            queue = newqueue
        return arr
    
    distances = [calc_dist(i,j) for i,j in machines]
    dists = dict()
    for i in range(len(machines)):
        for j in range(i+1,len(machines)):
            curdis = float('inf')
            for x,y in stones:
                if distances[i][x][y]==float('inf') or distances[j][x][y]==float('inf'):
                    continue
                curdis = min(curdis,distances[i][x][y]+distances[j][x][y])
            if curdis==float('inf'):
                return -1
            dists[(i,j)] = curdis
            dists[(j,i)] = curdis

    for i in range(len(machines)):
        dists[(i,target)] = distances[i][target[0]][target[1]]
        if dists[(i,target)]==float("inf"):
            return -1

    dp = dict()
    def dfs(index,visited):#最后一个是index，访问了visited的情况下的最小步数
        if (index,visited) in dp:
            return dp[(index,visited)]
        if visited+1==1<<len(machines):
            return dists[(index,target)]
        minimum = float('inf')
        for i in range(len(machines)):
            if visited & 1<<(len(machines)-i-1)==0:
                minimum = min(minimum,dists[(index,i)]+dfs(i,visited | 1<<(len(machines)-i-1)))

        dp[(index,visited)] = minimum
        return minimum

    return dfs(0,1<<(len(machines)-1))
minimalSteps(["S#O", "M..", "M.T"])
```

```python
# 801. Minimum Swaps To Make Sequences Increasing，两个数组最少需要交换多少次使得两个都严格递增
# 这个有点像股票交易费那样，换不换多种选择，保留两个dp
def minSwap(A, B):
    """
    :type A: List[int]
    :type B: List[int]
    :rtype: int
    """
    length = len(A)
    swap = [length for i in range(len(A))]# 当前位置交换情况下最少次数
    noswap = [length for i in range(len(A))]# 当前位置不交换情况下最少次数
    swap[0],noswap[0] = 1,0
    for i in range(1,len(A)):
        if A[i]>A[i-1] and B[i]>B[i-1]:
            swap[i] = swap[i-1]+1#当前需要交换，所以前面也必须交换
            noswap[i] = noswap[i-1]#当前不需要交换，所以前面也不需要交换
        if A[i]>B[i-1] and B[i]>A[i-1]:#如果交换以后也满足
            swap[i] = min(swap[i],noswap[i-1]+1)#当前需要交换后满足，所以前面不需要交换
            noswap[i] = min(noswap[i],swap[i-1]) #当前不交换,前面交换
    return min(swap[-1],noswap[-1])
minSwap(A = [1,3,5,4], B = [1,2,3,7])
```

```python
# 1478. Allocate Mailboxes，一个数组中插入k个位置，使得最小和最小
# dp[i][j] = dp[t][j-1]+CalDis(t+1,i), t从j-1到i-1，最后一个mailbox占据house数未知
def minDistance(houses, k):
    """
    :type houses: List[int]
    :type k: int
    :rtype: int
    """
    houses.sort()
    presum = [0]
    for i in houses:
        presum.append(presum[-1]+i)
    def calDis(i,j):#从i到j插入一个所需要的最小距离,右边减去左边
        idx1,idx2 = (i+j)//2,(i+j+1)//2
        return (presum[j+1]-presum[idx2]) - (presum[idx1+1]-presum[i])

    dp = [[float('inf') for _ in range(k)] for _ in range(len(houses))]
    for i in range(len(houses)):#house到i与mailbox到j所需要的最短距离
        for j in range(k):
            if j>=i:#如果mailbox比house更多
                dp[i][j] = 0
            elif j==0:#如果只有一个mailbox
                dp[i][j] = calDis(0,i)
            else:
                for t in range(j-1,i):#最后一个mailbox占据了后面一部分，更新距离
                    dp[i][j] = min(dp[i][j],dp[t][j-1]+calDis(t+1,i))
    return dp[-1][-1]
minDistance(houses = [1,4,8,10,20], k = 3)
```

```python
# 1477. Find Two Non-overlapping Sub-arrays Each With Target Sum,找出两个不覆盖的和为target的subarray，计算最少的长度和
# 建立presum，用来求当前的subarray，dp用来存储i以前的最短subarray长度
def minSumOfLengths(arr, target):
    """
    :type arr: List[int]
    :type target: int
    :rtype: int
    """
    presum = []
    for i in arr:
        if not presum:
            presum.append(i)
        else:
            presum.append(presum[-1]+i)
    dp = [float('inf') for _ in range(len(arr))]#i以前最短的的长度
    hashtable = {0:-1}
    res = float('inf')
    for i in range(len(arr)):
        if presum[i]-target in hashtable:
            curlen = i-hashtable[presum[i]-target]#当前长度
            if hashtable[presum[i]-target]==-1:#如果这是第一个满足的，无法更新
                dp[i] = curlen
            else:
                res = min(res,curlen+dp[hashtable[presum[i]-target]])#否则的话更新答案
                dp[i] = min(curlen,dp[i-1])#更新当前最短
        elif i>=1:
            dp[i] = dp[i-1]
        hashtable[presum[i]] = i
    return res if res!=float('inf') else -1
minSumOfLengths(arr = [3,2,2,4,3], target = 3)
```

```python
# 1569. Number of Ways to Reorder Array to Get Same BST
def numOfWays(nums):
    """
    :type nums: List[int]
    :rtype: int
    """
    mod = 10**9+7
    def fact(n,k):#得到组合数
        cur = 1
        for i in range(k+1,n+1):
            cur = cur*i//(i-k)
        return cur%mod

    dp = dict()
    def dfs(arr):#以arr的方法数，左边和右边可以随意穿插，只需要保持内部排序，相当于ltree+rtree长度里挑选ltree个无排序的位置，然后在内部排序
        if arr in dp:
            return dp[arr]
        if len(arr)==0:
            return 1
        ltree = []
        rtree = []
        for index in range(1,len(arr)):
            if arr[index]>arr[0]:
                rtree.append(arr[index]-arr[0])
            else:
                ltree.append(arr[index])
        left = dfs(tuple(ltree))
        right = dfs(tuple(rtree))
        cur = left*right*fact(len(ltree)+len(rtree),min(len(ltree),len(rtree)))%mod
        dp[arr] = cur
        return cur

    dfs(tuple(nums))
    return (dp[tuple(nums)]-1)%mod 
numOfWays(nums = [3,4,5,1,2])
```

```python
# 将m*n矩形用1*2铺满的方法
def dominoTiling(m,n):
    memo = {}
    def count_tilings_recursive(uncovered):
        if len(uncovered) & 1:
            return 0
        if not uncovered:
            return 1
        if uncovered not in memo:
            i, j = min(uncovered)
            memo[uncovered] = count_tilings_recursive(uncovered - {(i, j), (i, j + 1)}) + count_tilings_recursive(uncovered - {(i, j), (i + 1, j)})
        return memo[uncovered]


    def count_tilings(m, n):
        return count_tilings_recursive(frozenset((i, j) for i in range(max(m, n)) for j in range(min(m, n))))


    return count_tilings(m,n)
dominoTiling(22,3)
```

```python
# 1621. Number of Sets of K Non-Overlapping Line Segments，从n-1中选k段

def numberOfSets(n, k):
    """
    :type n: int
    :type k: int
    :rtype: int
    """
    mod = 10**9+7
    dp = dict()
    def dfs(n,k):
        if (n,k) in dp:
            return dp[(n,k)]
        if n==k:
            return 1
        if k==0:
            return 1
        if n<k or n==0:
            return 0
        
        res = 0
        res = 2*dfs(n-1,k)+dfs(n-1,k-1)-dfs(n-2,k)
        dp[(n,k)] = res%mod
        return dp[(n,k)]

    return dfs(n-1,k)%mod
numberOfSets(1000,500)
```

```python
# 数组轮流取头或尾，取k个数,给一个长度为k的数组B，然后取出来的k个数分别与数组B顺序相乘，求最大值
def getMaxProd(nums,arr,k):#从nums里左右选择k个数，与arr分别相乘的最大值
    dp = dict()
    def dfs(i,j):#左边是从0到i，右边是从j+1到结尾情况下的最大值
        if (i,j) in dp:
            return dp[(i,j)]
        if len(nums)-(j-i+1)==k:
            return 0
        curnum = len(nums)-(j-i+1)
        dp[(i,j)] = max(dfs(i+1,j)+nums[i]*arr[curnum],dfs(i,j-1)+nums[j]*arr[curnum])
        return dp[(i,j)]
    
    return dfs(0,len(nums)-1)
getMaxProd([1,5,2,3,4],[1,2,3],3)
```

```python
# 1240. Tiling a Rectangle with the Fewest Squares，用最少的正方形去构成一个矩形
# 三种切割方式，上下、左右切割成子问题，或者环形构成中间的一个
def tilingRectangle(n, m):
    """
    :type n: int
    :type m: int
    :rtype: int
    """
    dp = dict()
    def dfs(n,m):#(n,m)的矩阵需要的最少数目
        if n>m:
            return dfs(m,n)
        if (n,m) in dp:
            return dp[(n,m)]
        if n==m:
            return 1
        res = float('inf')
        for i in range(1,n):#上下切割
            res = min(res,dfs(i,m)+dfs(n-i,m))
        for i in range(1,m):#左右切割
            res = min(res,dfs(n,i)+dfs(n,m-i))
        # 四个长方形分别为(x,y),(n-x,z),(n-x-y+z,m-z),(x+y-z,m-y)
        for x in range(1,n):#四边切割，中间构成一个正方形
            for y in range(1,m):
                for z in range(1,y):
                    res = min(res,dfs(x,y)+dfs(n-x,z)+dfs(n-x-y+z,m-z)+dfs(x+y-z,m-y)+1)
        dp[(n,m)] = res
        return res

    return dfs(n,m)
tilingRectangle(11,13)
```
