```python
# 10. Regular Expression Matching，两个word的配对问题，.表示单一替换,*表示任意长度的前面字符
def isMatchRegular(s, p):
    """
    :type s: str
    :type p: str
    :rtype: bool
    """
    dp = [[False for _ in range(len(p)+1)] for _ in range(len(s)+1)]
    dp[0][0] = True
    for j in range(1,len(p)+1):
        if j%2==0 and dp[0][j-2] is True and p[j-1]=='*' and p[j-2]!='*':
            dp[0][j] = True
    for i in range(len(s)):
        for j in range(len(p)):
            if s[i]==p[j] or p[j]=='.':#当前符号相同
                dp[i+1][j+1] = dp[i][j]
            elif p[j]=='*' and j!=0:#要么这个通配符将前面一个消除掉，要么s[i]==p[j-1]将s的最后一个元素消掉，然后再进行
                dp[i+1][j+1] = dp[i+1][j-1]| (dp[i][j+1] &(s[i]==p[j-1] or p[j-1]=='.'))
    return dp[-1][-1]
isMatchRegular("aaa","ab*ac*a")

# 44. Wildcard Matching，两个word的配对问题，*表示任意长度通配，?表示单一替换
# 如果s[i] = p[j], dp[i+1][j+1] = dp[i][j]
# 如果不等，那么要么这个通配符不要，要么上面那个不要（通过从左到右来递进，避免重复运算）
def isMatchWildcard(s, p):
    """
    :type s: str
    :type p: str
    :rtype: bool
    """
    dp = [[False for _ in range(len(p)+1)] for _ in range(len(s)+1)]
    dp[0][0] = True
    for j in range(len(p)):
        if dp[0][j] is True and p[j] == '*':
            dp[0][j+1] = True
    for i in range(len(s)):
        for j in range(len(p)):
            if s[i]==p[j] or p[j] =='?':#当前符号相同
                dp[i+1][j+1] = dp[i][j]
            else:
                if p[j]=='*':#要么这个通配符不要，要么通配符抵消当前，继续保留通配符
                    dp[i+1][j+1] = dp[i+1][j] | dp[i][j+1]
    return dp[-1][-1]
isMatchWildcard('abc','a*')
```

```python
# 5 Longest Palindromic Substring: 最大回文子串    
def longestPalindrome(s):
    """
    :type s: str
    :rtype: str
    """
    def PalindromeSeg(s,left,right):
        while left>=0 and right<len(s) and s[left]==s[right]:
            left = left-1
            right = right+1
        return left+1,right-1

    if not s:
        return ""
    oddlength = 0
    # odd
    for i in range(len(s)):
        left,right = PalindromeSeg(s,i,i)
        if right-left+1>oddlength:
            oddlength = right-left+1
            start,end = left,right
    length = end-start+1
    for i in range(len(s)-1):
        left,right = PalindromeSeg(s,i,i+1)
        if right-left+1>length:
            length = right-left+1
            start,end = left,right
    return s[start:end+1]  
```

```python
# 最长公共子串
# 718. Maximum Length of Repeated Subarray
def longestCommonString(s,p):
    dp = [[0 for _ in range(len(p)+1)] for _ in range(len(s)+1)]
    for i in range(len(s)):
        for j in range(len(p)):
            if s[i]==p[j]:
                dp[i+1][j+1] = dp[i][j]+1
    return max([max(i) for i in dp])

# 1143. Longest Common Subsequence，最长公共子序列
def longestCommonSubsequence(text1, text2):
    """
    :type text1: str
    :type text2: str
    :rtype: int
    """
    m = len(text1)
    n = len(text2)
    if m==0 or n==0:
        return 0
    dp = [[0 for _ in range(n)] for _ in range(m)]
    dp[0][0] = text1[0]==text2[0]
    for i in range(m):
        for j in range(n):
            if i==0:
                dp[i][j] = dp[i][j-1] or text1[i]==text2[j]
            elif j==0:
                dp[i][j] = dp[i-1][j] or text1[i]==text2[j]                
            else:
                if text1[i] == text2[j]:
                    dp[i][j] = dp[i-1][j-1]+1
                else:
                    dp[i][j] = max(dp[i-1][j], dp[i][j-1])
    return dp[-1][-1]*1#需要乘以1处理boolean的问题
longestCommonSubsequence(text1 = "bc", text2 = "a" )

# 583. Delete Operation for Two Strings,删除最少的char，使得两个word相同，其实就是求最长公共子序列
def minDistance(word1, word2):
    """
    :type word1: str
    :type word2: str
    :rtype: int
    """
    dp = [[0 for _ in range(len(word2)+1)] for _ in range(len(word1)+1)]
    for i in range(len(word1)+1):
        dp[i][0] = i
    for j in range(len(word2)+1):
        dp[0][j] = j
    for i in range(len(word1)):
        for j in range(len(word2)):
            if word1[i]==word2[j]:
                dp[i+1][j+1] = dp[i][j]
            else:
                dp[i+1][j+1] = min(dp[i+1][j],dp[i][j+1])+1
    return dp[-1][-1]

minDistance("eadfega", "eatefadf")

# 最长公共子序列
def LCS_length(X,Y):
    """
    the largest common sequence doesn't need to be continuous.
    X: str
    Y: str
"""
    m=len(X)
    n=len(Y)
    b=[["" for i in range(n+1)] for i in range(m+1)]
    c=[[0 for _ in range(n+1)] for _ in range(m+1)]
    for i in range(0,m):
        for j in range(0,n):
            if X[i]==Y[j]:
                c[i+1][j+1]=c[i][j]+1
                b[i+1][j+1]='ul'
            elif c[i][j+1]>=c[i+1][j]:
                c[i+1][j+1]=c[i][j+1]
                b[i+1][j+1]='up'
            else:
                c[i+1][j+1]=c[i+1][j]
                b[i+1][j+1]='left'
    return c,b

def Print_LCS(b,X,i,j):
    if i==0 or j==0:
        return
    if b[i][j]=='ul':
        Print_LCS(b,X,i-1,j-1)
        print(X[i-1]) # X starts from 0
    elif b[i][j]=='left':
        Print_LCS(b,X,i-1,j)
    elif b[i][j]=='up':
        Print_LCS(b,X,i,j-1)
        
X='ABCBDAB'
Y='BDCABA'
c,b=LCS_length(X,Y)
# Print_LCS(b,X,len(X),len(Y))

# 1062. Longest Repeating Substring
def longestRepeatingSubstring(S):
    """
    :type S: str
    :rtype: int
    """
    dp = [[0 for _ in range(len(S))] for _ in range(len(S))]
    for i in range(len(S)):
        for j in range(i+1,len(S)):
            if S[i]==S[j]:
                if i==0:
                    dp[i][j]=1
                else:
                    dp[i][j] = dp[i-1][j-1]+1
    return max(max(i) for i in dp)
longestRepeatingSubstring("aaaa")

# 516. Longest Palindromic Subsequence
def longestPalindromeSubseq(s):
    """
    :type s: str
    :rtype: int
    """
    t = s[::-1]
    m = len(s)
    if s==t:
        return m
    dp = [[0 for _ in range(m+1)] for _ in range(m+1)]      
    for i in range(m):
        for j in range(0,m):
            if s[i] == t[j]:
                dp[i+1][j+1] = dp[i][j]+1
            else:
                dp[i+1][j+1] = max(dp[i+1][j], dp[i][j+1])        
    return dp[-1][-1]

# 1312. Minimum Insertion Steps to Make a String Palindrome,插入多少次使得回文，也就是找最长的回文子串
def minInsertions(s):
    n = len(s)
    dp = [[0] * (n + 1) for i in range(n + 1)]
    for i in range(n):
        for j in range(n):
            dp[i + 1][j + 1] = dp[i][j] + 1 if s[i] == s[~j] else max(dp[i][j + 1], dp[i + 1][j])
    return n - dp[n][n]
minInsertions(s = "leetcode")
```

```python
# 1092. Shortest Common Supersequence，最短公共超级序列=总长度-最长公共子序列
def shortestCommonSupersequence(A, B):
    def lcs(A, B):
        n, m = len(A), len(B)
        dp = [["" for _ in range(m + 1)] for _ in range(n + 1)]#长度增加1用来处理0的问题，不需要设立两个维度
        for i in range(n):
            for j in range(m):
                if A[i] == B[j]:
                    dp[i + 1][j + 1] = dp[i][j] + A[i]
                else:
                    dp[i + 1][j + 1] = max(dp[i + 1][j], dp[i][j + 1], key=len)
        return dp[-1][-1]

    res, i, j = "", 0, 0
    for c in lcs(A, B):
        while A[i] != c:
            res += A[i]
            i += 1
        while B[j] != c:
            res += B[j]
            j += 1
        res += c
        i, j = i + 1, j + 1
    return res + A[i:] + B[j:]
shortestCommonSupersequence(A= "abac", B= "cab")
```

```python
# 712. Minimum ASCII Delete Sum for Two Strings,两个字符串，通过删减而相同，求最小删减值
def minimumDeleteSum(s1, s2):
    """
    :type s1: str
    :type s2: str
    :rtype: int
    """
    # dp[i][j] means the minimum cost for s1[:i] and s2[:j]
    # if s1[i]==s2[j]: dp[i][j] = dp[i-1][j-1]
    # if s1[i]!=s2[j]: dp[i][j] = dp[i-1][j]+s1[i],dp[i][j-1]+s2[j]
    dp = [[0 for _ in range(len(s2)+1)] for _ in range(len(s1)+1)]
    for i in range(len(s1)):
        dp[i+1][0] = dp[i][0]+ord(s1[i])
    for j in range(len(s2)):
        dp[0][j+1] = dp[0][j]+ord(s2[j])
    for i in range(len(s1)):
        for j in range(len(s2)):
            if s1[i]==s2[j]:
                dp[i+1][j+1] = dp[i][j]
            else:
                dp[i+1][j+1] = min(dp[i][j+1]+ord(s1[i]),dp[i+1][j]+ord(s2[j]))
    return dp[-1][-1]
minimumDeleteSum(s1 = "delete", s2 = "leet")

# 72. Edit Distance,两个word通过替换/插入/删除变成一致，和最长子序列有点像
def minDistance(word1, word2):
    """
    :type word1: str
    :type word2: str
    :rtype: int
    """
    m,n = len(word1),len(word2)
    dp = [[0 for _ in range(n+1)] for _ in range(m+1)]#从长度为0到整个长度
    for i in range(m+1):
        dp[i][0] = i
    for j in range(n+1):
        dp[0][j] = j
    for i in range(m):
        for j in range(n):
            if word1[i]==word2[j]:
                dp[i+1][j+1] = dp[i][j]
            else:
                # dp[i][j] is for replacing
                # dp[i+1][j] is for inserting
                # dp[i][j+1] is for deleting
                dp[i+1][j+1] = min(dp[i][j],dp[i][j+1],dp[i+1][j])+1
    return dp[-1][-1]
```

```python
# 97. Interleaving String, 判断s3是否由s1,s2交错构成
# 当前(i+1,j+1)，由上一位&对应元素相等构成
def isInterleave(s1, s2, s3):
    """
    :type s1: str
    :type s2: str
    :type s3: str
    :rtype: bool
    """
    if len(s1)+len(s2)!=len(s3):
        return False
    dp = [[False for _ in range(len(s2)+1)] for _ in range(len(s1)+1)]
    dp[0][0] = True
    for i in range(len(s1)):
        dp[i+1][0] = dp[i][0] & (s1[i] == s3[i])
    for j in range(len(s2)):
        dp[0][j+1] = dp[0][j] & (s2[j] == s3[j])
    for i in range(len(s1)):
        for j in range(len(s2)):
            dp[i+1][j+1] = (dp[i][j+1] & (s1[i]==s3[i+j+1])) or (dp[i+1][j] & (s2[j]==s3[i+j+1]))
    return dp[-1][-1]
isInterleave(s1 = "aabcc", s2 = "dbbca", s3 = "aadbbcbcac")
```

```python
# 115. Distinct Subsequences, s中与t相同的子串的个数
def numDistinct(s, t):
    """
    :type s: str
    :type t: str
    :rtype: int
    """
    if not t:
        return 1
    if not s:
        return 0
    dp = [[0 for j in range(len(s))] for _ in range(len(t))]#以t[i],s[j]结尾的部分最大的能包括t[i]所有元素的子序列个数
    dp[0][0] = 1 if s[0]==t[0] else 0
    for j in range(1,len(s)):
        dp[0][j] = dp[0][j-1]+(s[j]==t[0])

    for i in range(1,len(t)):
        for j in range(i,len(s)):
            if t[i]!=s[j]:#如果两个不相等
                dp[i][j] = dp[i][j-1]
            else:#相等的话，两个退一步+source退一步得到的
                dp[i][j] = dp[i][j-1]+dp[i-1][j-1]
    return dp[-1][-1]

numDistinct(s = "rabbbit", t = "rabbit")
```

```python
# 1035. Uncrossed Lines
def maxUncrossedLines(A, B):
    """
    :type A: List[int]
    :type B: List[int]
    :rtype: int
    """
    dp = [[0 for _ in range(len(B)+1)]for _ in range(len(A)+1)]
    for i in range(len(A)):
        for j in range(len(B)):
            if A[i]==B[j]:
                dp[i+1][j+1]= dp[i][j]+1
            else:
                dp[i+1][j+1] = max(dp[i+1][j],dp[i][j+1])
    return dp[-1][-1]

maxUncrossedLines(A = [2,5,1,2,5], B = [10,5,2,1,5,2])
```

```python
# 1420. Build Array Where You Can Find The Maximum Exactly K Comparisons
# n个数，每个数在1-m之间，新高次数为k，总共有多少种方法
# 一个基本的思路是dp[k,n] = dp[k,n-1]+dp[k-1,n-1]上面进行修改
def numOfArrays(n, m, k):
    """
    :type n: int
    :type m: int
    :type k: int
    :rtype: int
    """
    # search_cost是指每次遇到新的更大的数，就要+1
    # n个数，每个数在1-m之间，新高次数为k
    mod = 10**9+7
    if k==0:
        return 0
    dp = [[[0 for _ in range(m+1)] for _ in range(n+1)] for _ in range(k+1)]
    for j in range(1,n+1):
        for m0 in range(1,m+1):#长度为j，最大值为m0
            dp[1][j][m0]=m0**(j-1)
    for i in range(2,k+1):#新高次数为2~k
        for j in range(i,n+1):#新高次数为i，长度最小为i,最大为n
            for m0 in range(i,m+1):#长度为j，新高为i,最大值为m
                temp =0
                for t in range(i-1,m0):
                    temp+=dp[i-1][j-1][t]
                dp[i][j][m0] = (dp[i][j-1][m0]*m0+temp)%mod
    return sum(dp[k][n])%mod
numOfArrays(n = 50, m = 100, k = 7)
```
